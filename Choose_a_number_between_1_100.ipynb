```python
import time
import numpy as np
from IPython.display import display, clear_output
import ipywidgets as widgets

# Parameters
number_range = 100  # Range of numbers (1 to 100)
total_duration = 10  # Total duration in seconds
frames = total_duration * 10  # Total number of frames in the animation (10 frames per second)
interval = 0.1  # Time between frames in seconds
slowdown_frames = 30  # Number of frames to slow down

# Global variables
selected_number = np.random.randint(1, number_range + 1)
current_number = 1
correct_guesses = 0  # Track the number of correct guesses

# Function to animate the number cycling
def animate_numbers(guess):
    global selected_number, current_number, correct_guesses
    selected_number = np.random.randint(1, number_range + 1)
    current_number = 1

    for frame in range(frames):
        if frame < frames - slowdown_frames:
            current_number = (frame % number_range) + 1
        else:
            slowdown_factor = (frame - (frames - slowdown_frames)) / slowdown_frames
            if frame % int((1 - slowdown_factor) * 10 + 1) == 0:
                current_number = (current_number % number_range) + 1

            if frame == frames - 1:
                current_number = selected_number

        clear_output(wait=True)
        display(widgets.HTML(f"<h1 style='font-size: 24px; color: black;'>Current number: {current_number}</h1>"))
        time.sleep(interval)

    # Display the result after the animation completes
    if guess == selected_number:
        result = f"Congratulations! Your guess {guess} was correct!"
        correct_guesses += 1
    else:
        result = f"Sorry, your guess {guess} was wrong. The correct number was {selected_number}."

    # Display the result and the number of correct guesses
    display_result(result)

# Function to display the result
def display_result(result):
    result_label = widgets.Label(value=result)
    correct_label = widgets.Label(value=f"Number of correct guesses: {correct_guesses}")
    display(result_label, correct_label)

# Function to start the game
def start_game(b):
    clear_output(wait=True)
    display(guess_input, start_button)
    animate_numbers(guess_input.value)
    display(play_again_button, quit_button)

# Function to play again
def play_again(b):
    clear_output(wait=True)
    display(guess_input, start_button)

# Function to quit the game
def quit_game(b):
    clear_output(wait=True)
    display(widgets.Label(value="Thank you for playing!"))

# Create widgets for user input
guess_input = widgets.BoundedIntText(value=1, description='Your Guess:', min=1, max=100)
start_button = widgets.Button(description='Start Game')
play_again_button = widgets.Button(description='Play Again')
quit_button = widgets.Button(description='Quit')

# Assign button click functions
start_button.on_click(start_game)
play_again_button.on_click(play_again)
quit_button.on_click(quit_game)

# Display the input widgets
display(guess_input, start_button)
```
